```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
ndata["Location"]
```

```python
# Vectorization

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(ndata["Location"])
```

```python
# Cosine Similarity

cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)
cosine_sim
```

```python
ind_job =  pd.Series(ndata["Job_Name"]).drop_duplicates()
#indices["Data Science"]
indices1 = pd.Series(ind_job.index,index=ind_job.values)
indices1.index
```

```python
indices1['Data Analyst']
```

```python
ind_job.values
```

```python
indices = pd.Series(ndata.index,index=ndata["Job_Name"]).drop_duplicates()
indices.index
```

```python
indices
```

```python
indices["Data Science"]
```

```python
def get_recommendations(title, cosine_sim = cosine_sim):
    idx = indices[title]
    sim_scores = enumerate(cosine_sim[idx])       
    sim_scores = sorted(sim_scores,key=lambda x:x[1],reverse =True)
    sim_scores = sim_scores[1:11] 
    
    sim_index = [i[0] for i in sim_scores]   
    print(ndata["Job_Name"].iloc[(sim_index)])
    print(ndata["Location"].iloc[(sim_index)])
```

```python
print("Top 10 recommendations based on location are :")
print("\n")
get_recommendations("Data Controller")
```
